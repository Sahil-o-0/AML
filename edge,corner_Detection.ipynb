```python
import sys
import cv2
import numpy as np
input_file = sys.argv[1]
img = cv2.imread('/content/OIP.jpg', cv2.IMREAD_GRAYSCALE)
h, w = img.shape
sobel_horizontal = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=5)
sobel_vertical = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=5)
laplacian = cv2.Laplacian(img, cv2.CV_64F)
canny = cv2.Canny(img, 50, 240)
from google.colab.patches import cv2_imshow
cv2_imshow(img)
cv2_imshow(sobel_horizontal)
cv2_imshow(sobel_vertical)
cv2_imshow(laplacian)
cv2_imshow(canny)
```

```python
import sys
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
input_file = sys.argv[1]
img = cv2.imread('/content/OIP.jpg')
cv2_imshow(img)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_gray = np.float32(img_gray)
img_harris = cv2.cornerHarris(img_gray, 7, 5, 0.04)
img_harris = cv2.dilate(img_harris, None)
img[img_harris > 0.01 * img_harris.max()] = [0, 0, 0]
cv2_imshow(img)

```
